```
data=pd.read_csv("headlines.csv",encoding="ISO-8859-1")
list=data['headline'].tolist()
sid = SentimentIntensityAnalyzer()
data['headline'] = data.headline.map(lambda x: str(x).lower().translate(str.maketrans('','', string.punctuation)))
data['headline']=data.headline.astype("str")
data['publish_year'] = data.date.str[8:12]
data['publish_month'] = data.date.str[:4]
data['publish_date']=data['publish_year'] +"-"+ data['publish_month']
sid = SentimentIntensityAnalyzer()
label=[]
for sentence in list:
     print(sentence)
     ss = sid.polarity_scores(sentence)
     for k in sorted(ss):
         
        print('{0}: {1}, '.format(k, ss[k]), end='')
     label.append(ss)
     print()   
data['label'] =label 
x=[]
for i in range(794):
  d=max(data.label[i]['neu'],data.label[i]['neg'],data.label[i]['pos'])
  if data.label[i]['neu'] ==d:
    x.append("0")
  elif data.label[i]["pos"]==d:
    x.append(1)
  elif data.label[i]["neg"]==d:
    x.append(-1)
data["LABEL"]=x
```
